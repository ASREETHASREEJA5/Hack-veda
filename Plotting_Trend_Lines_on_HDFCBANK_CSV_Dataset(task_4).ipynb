```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/StockDataHDFCBANK.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df = df[['Date', 'Close']]

# Calculate trend lines
# For simplicity, we'll manually specify points for trend lines in this example
# In practice, you might use algorithms to identify these points
uptrend_points = df[['Date', 'Close']].iloc[[0, 15, -2]]  # Sample points for uptrend line
downtrend_points = df[['Date', 'Close']].iloc[[1, -20, -15]]  # Sample points for downtrend line

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], label='Close Price', marker='o')

# Plot trend lines
plt.plot(uptrend_points['Date'], uptrend_points['Close'], label='Uptrend Line', color='green')
plt.plot(downtrend_points['Date'], downtrend_points['Close'], label='Downtrend Line', color='red')

# Customize the plot
plt.title('Stock Price with Trend Lines')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
